W automatycznym systemie analizy obraz贸w (dziaanie bez nadzoru operatora) konieczne jest zastosowanie metody binaryzacji, kt贸ra w spos贸b automatyczny wyznacza pr贸g binaryzacji.

#### Iteracyjne wyznaczenie progu

Jednym z najprostszych podej jest iteracyjna procedura wyliczania progu.
Jako pierwsze przybli偶enie progu ($k$) przyjmuje si rednia jasno na obrazie.
Nastpnie, na podstawie $k$,  dzieli si obraz na dwa podobrazy $I_0$ i  $I_1$ (dwie klasy $C_0$ i $C_1$).
Dla ka偶dego z nich oblicza si redni jasno: $m_0$ i $m_1$.
Jako nowy pr贸g przyjmuje si:
\begin{equation}
\label{eq:thresholding:mean_auto} \tag{1}
k_{new} = \frac{m_0 + m_1}{2}
\end{equation}
Procedur kontynuuje si do momentu, a偶 r贸偶nica pomidzy dwoma kolejnymi progami bdzie mniejsza ni偶 zadana warto.


**Zadanie: zaimplementowa opisany powy偶ej algorytm.**


Jak mo偶na zauwa偶y, do poprawnego dziaania metody potrzebne bdzie obliczanie redniej jasnoci, r贸wnie偶 dla pewnych podobraz贸w.
Wykorzystamy do tego znormalizowanych histogram:
\begin{equation}
\label{eq:p_i} \tag{2}
p_i = n_i/N,   \sum_{i=0}^L p_i = 1
\end{equation}
gdzie: $n_i$ liczba pikseli o jasnoci $i$ ($i = 0,1, ... L-1$) - histogram, $L$ - liczba poziom贸w jasnoci, $N$ - liczba pikseli na obrazie ($N = n_0 + n_1 + ... + n_{L-1}$).

Jeli podzielimy obraz na dwie klasy $C_0$ i $C_1$ (to i obiekty albo obiekty i to) z progiem podziau oznaczonym jako $k$, to do klasy $C_0$ nale偶e bd piksele o poziomach $[0,k]$, a do klasy $C1$ piksele o poziomach $[k+1,L-1]$.

Wtedy prawdopodobiestwo, 偶e piksel nale偶y do klasy $C_0$ wynosi:
\begin{equation}
\label{eq:P_0} \tag{3}
P_0(k) = \sum_{i=0}^{k} p_i
\end{equation}

Podobnie prawdopodobiestwo, 偶e nale偶y do klasy $C_1$ wynosi:

\begin{equation}
\label{eq:P_1} \tag{4}
P_1(k) = \sum_{i=k+1}^{L-1} p_i = 1 - P_0(k)
\end{equation}

redni jasno pikseli nale偶cych do klasy $C_0$ mo偶na wyznaczy na podstawie:

\begin{equation}
\label{eq:m_0} \tag{5}
m_0(k) = \sum_{i=0}^{k} iP(i|C_0)
\end{equation}

gdzie: $|$ oznacza prawdopodobiestwo warunkowe, a wyraz $P(i|C_0)$ - prawdopodobiestwo dla wartoci $i$ pod warunkiem, 偶e $i$ nale偶y do klasy $C_0$.
R贸wnanie to jest szczeg贸lnym przypadkiem wykorzystania moment贸w statystycznych do wyliczania pewnych parametr贸w statystycznych - w tym przypadku redniej.

Wykorzystujc regu Bayesa:

\begin{equation}
\label{eq:P_A_B} \tag{6}
P(A|B) = P(B|A)P(A)/P(B)
\end{equation}
mo偶emy zapisa:

\begin{equation}
\label{eq:m_0k} \tag{7}
m_0(k) = \sum_{i=0}^{k} i P(C_0|i)P(i)/P(C_0)
\end{equation}
Wyraz $P(C_0|i) = 1$, gdy偶 z zao偶enia rozpatrujemy tylko piksele nale偶ce do klasy $C_0$.
Wyraz $P(i)$ stanowi $i$-ty element znormalizowanego histogramu tj. $P(i) = p_i$, a $P(C_0)$ to prawdopodobiestwo przynale偶noci do klasy $C_0$ okrelone wczeniej $P(C_0) = P_0(k)$.
Ostatecznie mo偶emy wic zapisa:

\begin{equation}
\label{eq:thresholding:meanHistC0} \tag{8}
m_0(k) = \frac{1}{P_0(k)} \sum_{i=0}^{k} i p_i
\end{equation}

Na podstawie analogicznych rozwa偶a mo偶na wyprowadzi wz贸r na redni jasnoci pikseli nale偶cych do klasy $C_1$:
\begin{equation}
\label{eq:thresholding:meanHistC1}  \tag{9}
m_1(k) = \frac{1}{P_1(k)} \sum_{i=k+1}^{L-1} i p_i
\end{equation}

rednia jasno caego obrazu dana jest zale偶noci:
\begin{equation}
\label{eq:thresholding:meanHistGlobal} \tag{10}
m_G = \sum_{i=0}^{L-1} ip_i
\end{equation}

#### Metoda Otsu

Jednym z czciej wykorzystywanych algorytm贸w wyznaczania progu jest metoda zaproponowana w roku 1979 przez Nobuyuki Otsu w artykule pt. "A Threshold Selection Method from Gray-Level Histograms" (mo偶na odszuka na IEEE Xplore).
W algorytmie zakada si, 偶e obraz zawiera piksele nale偶ce do dw贸ch klas (obiekt贸w i ta) tj. histogram obrazu jest bi-modalny (ma dwa maksima).
Pr贸g podziau obliczany jest tak, aby wariancja midzyklasowa bya maksymalna.
W tym sensie metod Otsu mo偶na nazwa optymaln.

Wprowad藕my teraz wska藕nik "jakoci" wybranego progu podziau $k$, kt贸ry bdziemy optymalizowa.
W algorytmie Otsu jest to:

\begin{equation}
\label{eq:thresholding:otsu:eta} \tag{11}
\eta(k) = \frac{\sigma^2_B(k)}{\sigma^2_G}
\end{equation}
gdzie:  $\sigma^2_G$ - wariancja globalna, kt贸ra mo偶e zosta obliczona na podstawie moment贸w statystycznych jako:

\begin{equation}
\label{eq:sigma} \tag{12}
\sigma^2_G =  \sum_{i=0}^{L-1} (i - m_G)^2 p_i
\end{equation}
a $\sigma^2_B$ jest wariancj midzyklasow, kt贸ra jest zdefiniowana jako:
\begin{equation}
\label{eq:sigma_b} \tag{13}
\sigma^2_B(k) =  P_0(k)(m_0(k) - m_G)^2 + P_1(k)(m_1(k) - m_G)^2
\end{equation}
R贸wnianie to mo偶na r贸wnie偶 przeksztaci do:
\begin{equation}
\label{eq:thresholding:otsu:sigma_B_2} \tag{14}
\sigma^2_B(k) =  P_0(k)P_1(k)(m_0(k) - m_1(k))^2 = \frac{(m_G P_0(k) - m(k) )^2}{P_0(k)(1-P_0(k))}
\end{equation}
gdzie:
\begin{equation}
\label{eq:mk} \tag{15}
m(k) = \sum_{i=0}^{k} i p_i
\end{equation}

Taki zapis pozwala przyspieszy obliczenia.
Warto $m_G$ wyznaczana jest jednokrotnie, a zachodzi tylko potrzeba obliczania $m(k)$ i $P_0(k)$ w ka偶dej iteracji.
Warto te偶 zwr贸ci uwag, 偶e r贸wnanie ma sens dla $P_0 > 0$.

Warto zauwa偶y, 偶e z postaci r贸wnania (\ref{eq:thresholding:otsu:sigma_B_2}) wynika, 偶e im wiksza odlego pomidzy rednimi $m_0$ i $m_1$ tym warto wariancji midzyklasowej jest wiksza.
Pokazuje to, 偶e przyjty wsp贸czynniki mo偶e by podstaw do separacji dw贸ch klas - im jego warto jest wiksza, tym lepsze rozdzielenie.
Dodatkowo, z r贸wnania (\ref{eq:thresholding:otsu:eta}) wynika, 偶e $\eta(k)$ zale偶y tylko od wariancji midzyklasowej $\sigma^2_B(k)$, gdy偶 wariancja globalna $\sigma^2_G$ jest staa.
Zatem w procesie optymalizacji nale偶y d偶y do maksymalizacji wska藕nika $\eta$.

Nale偶y te偶 pamita, 偶e wsp贸czynnik jest poprawnie okrelony tylko dla wartoci $\sigma^2_G > 0$.
Przy czym, warto $0$ mo偶e on przyj tylko dla obrazu o jednym poziomie szaroci - w takim przypadku trudno m贸wi o podziale pikseli na dwie klasy (skoro wystpuje tylko jedna).

Ostatecznie optymalny pr贸g binaryzacji $\bar{k}$ wyliczamy na podstawie zale偶noci:
\begin{equation}
\label{eq:sigma_B_k} \tag{16}
\sigma^2_B(\bar{k}) \max\limits_{l \in[0,L-1]} {\sigma^2_B(k) }
\end{equation}

Uwagi:
- mo偶e si zdarzy, 偶e znajdziemy wicej ni偶 jedno maksimum tj. wicej wartoci $\bar{k}$.
  W takim przypadku zwykle zakada si, 偶e pr贸g bdzie redni otrzymanych wartoci.
- liczby $P_0(\bar{k})$ i $P_1(\bar{k})$ odpowiadaj powierzchni zajmowanej przez obiekty klas $C_0$ i $C_1$.
- liczby $m_0(\bar{k})$ i $m_1(\bar{k})$ odpowiadaj redniej jasnoci obiekt贸w klas $C_0$ i $C_1$.
- warto parametru $\eta(\bar{k})$ okrela "jako" wyznaczonego progu -- im wiksza tym lepiej.

```python
#2 Wyznaczenie histogramu znormalizowanego oraz obliczenie jasnoci
histogram1=cv2.calcHist([coins],[0],None,[256],[0,256])
histogram1=np.squeeze(histogram1)
histogram_cumsum=histogram1.cumsum()


hist_cumsum_normalized = cv2.normalize(histogram_cumsum,None,0,histogram1.max())

hist_cumsum_normalized = (histogram_cumsum/(coins.shape[0]*coins.shape[1]))
hist_normalized = (histogram1/(coins.shape[0]*coins.shape[1]))

#wyliczenie rednej
coins_mean=int(np.round(np.dot(np.arange(256),hist_normalized)))

#3 Zdefiniowanie 256-elementowego wektora na wsp贸czynnik sigma^B
sigma_B=np.zeros(256)
n_k=np.zeros(256)

sigma_G=np.dot(((np.arange(256)-coins_mean)**2),hist_normalized)

#4 Wyznacznie w petli sigma^b(k)
for k in range(0,256):
    p0=hist_cumsum_normalized[k]
    if p0 > 0 and p0 < 1:
        m_k=np.dot(np.arange(k+1),hist_normalized[0:k+1])
        sigma_B[k]=(((coins_mean*p0-m_k)**2)/(p0*(1-p0)))
        n_k[k]=(sigma_B[k]/sigma_G)

#5 Wywietlenie przebiegu sigma^B(k)
plt.plot(sigma_B)
plt.grid
plt.title("Przebieg (sigma_B)^2")
plt.show()
```

```python
# 6 Wyznaczenie wartoci    dla kt贸rej wsp贸czynnik  (sigma_b)^2 jest maksymalny. 
sigma_maximum = 0

for i in range(len(sigma_B)):
    if(sigma_B[i]>sigma_maximum):
        sigma_maximum = sigma_B[i]
        k_max=i

print("Pr贸g:",k_max)
```

### Binaryzacja lokalna


Analiza wynik贸w binaryzacji dla obraz贸w _rice.png_ i _catalogue.png_ pokazuje, 偶e globalna binaryzacja nie najlepiej dziaa dla obraz贸w o niejednorodnym owietleniu.
Dla obu obraz贸w trudno r贸wnie偶 wyznaczy odpowiedni pr贸g "rcznie".

Metod, kt贸ra pozwala poprawi wyniki binaryzacji, jest binaryzacja lokalna (niekiedy zwana adaptacyjn).
W jednym z wariant贸w polega ona na wyznaczeniu progu osobno dla ka偶dego piksela na podstawie jego otoczenia (tj. wasnoci jego kontekstu, okna).

```python
import numpy as np
def binary(obraz, W):
    (X, Y) = obraz.shape
    bin_obraz = np.zeros(obraz.shape)
   
    for j in range(int(W/2), Y-int(W/2)):
        for i in range (int(W/2), X-int(W/2)):
            otoczenie = obraz[i-int(W/2):i+int(W/2),j-int(W/2):j+int(W/2)]
            srednia = np.mean(otoczenie)
            bin_obraz[i][j] = obraz[i][j] > srednia

    
    fig, (ax1,ax2) = plt.subplots(1, 2,figsize=(10,10))


    ax1.imshow(obraz, 'gray', vmin=0, vmax=256)
    ax1.axis('off')
    ax2.imshow(bin_obraz, 'gray')
    ax2.axis('off')

#dla rice.png
rice = cv2.imread("rice.png", cv2.IMREAD_GRAYSCALE)    
binary(rice, 10)    
```

```python
def Sauvol(obraz,W,R,k):
    (X, Y) = obraz.shape
    bin_obraz = np.zeros(obraz.shape)

    for j in range(int(W/2), Y-int(W/2)):
        for i in range (int(W/2), X-int(W/2)):
            otoczenie = obraz[i-int(W/2):i+int(W/2),j-int(W/2):j+int(W/2)]
            srednia = np.mean(otoczenie)*(1-k*((np.std(otoczenie)/R)-1))
            bin_obraz[i][j] = obraz[i][j] > srednia
          
    
    fig, (ax1,ax2) = plt.subplots(1, 2,figsize=(10,10))


    ax1.imshow(obraz, 'gray', vmin=0, vmax=256)
    ax1.axis('off')
    ax2.imshow(bin_obraz, 'gray')
    ax2.axis('off')
#dla rice.png
rice = cv2.imread("rice.png", cv2.IMREAD_GRAYSCALE)    
Sauvol(rice, 15,128,0.15) 
```
